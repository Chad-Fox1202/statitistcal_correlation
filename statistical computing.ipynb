```python
import pandas as pd

df = pd.read_csv('tj.csv')

df2 = pd.read_csv('no_tj_differences.csv')

```

```python
age=df['age before surgery ']
ageNoTJ=df2['player_age']

recov=df['recovery_time']

surgeon=df['surgeon']

andrews=df['Andrews x ']

neal=df['Neal']

kpercent=df['k_percentage diff']
kpercentNoTJ = df2['k_percentage diff']

bbpercent=df['bb_percentage diff']
bbpercentNoTJ=df2['bb_percentage diff']

exitvelo=df['avg_exit_velocity diff']
exitveloNpTJ=df2['avg_exit_velocity diff']

hardhit=df['hard_hit_percentage diff']
hardhitNoTJ=df2['hard_hit_percentage diff']

fastvelo=df['ff_velo diff']
fastveloNoTJ=df2['ff_velo diff']

fastrpm=df['ff_rpm diff']
fastrpmNoTJ=df2['ff_rpm diff']

era=df['era diff']
eraNoTJ=df2['era diff ']

whip=df['whip diff ']
whipNoTJ=df2['whip diff']
```

```python
import statsmodels.api as sm
```

```python
#regression models with surgeon
X=pd.DataFrame({'andrews':andrews,'neal':neal})
y=pd.DataFrame({'recov':recov})
model1=sm.OLS(y,X)
results=model1.fit()
print(results.summary())
```

```python
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
from scipy.stats import ttest_ind, norm
import seaborn as sns
```

```python
ttestERA = ttest_ind(era,eraNoTJ)
ttestERA
```

```python
sns.distplot(era, color='r', label = 'ERA of TJ pitchers')
sns.distplot(eraNoTJ, color='b', label = 'ERA of non TJ pitchers')
plt.xlabel('ERA difference')
plt.ylabel('probability density')
plt.title('ERA difference PDF')
plt.legend()
plt.show
```
